```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
alpha = 0.01
R = np.arange(1, 30+.1, .1)
n_gen = 300
eta = np.zeros((len(R), n_gen))
eta_start = 900
eta[:,0] = eta_start
```

```python
def model():
    for t in range(n_gen-1):
        eta[:, t+1] = np.transpose(R)*eta[:, t]*np.exp(-alpha*eta[:, t])
    return eta
```

```python
eta = model()
plt.plot(R, eta[:,-100:], '.')
plt.xlabel("R")
plt.ylabel(r"$\tau$")
plt.show()
```

```python
R = np.arange(1, 30+.1, .1)
R_four = [40, 90, 120, 220]
print(R[R_four])

for r in R_four:
    plt.plot(eta[r,:41], label=f'R={R[r]:.1f}')

plt.legend()
plt.title('Task 3b) Four dynamics of population')
plt.xlabel(r'$\tau$')
plt.ylabel(r'$\eta_{\tau}$')
plt.show()
```
