```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
alpha = 0.01
R = np.arange(1, 30+.1, .1)
n_gen = 300

eta = np.zeros((len(R), n_gen))
eta_start = 900
eta[:,0] = eta_start
```

```python
def model():
    for t in range(n_gen-1):
        eta[:, t+1] = np.transpose(R)*eta[:, t]*np.exp(-alpha*eta[:, t])
    return eta
```

```python
eta = model()
plt.plot(R, eta[:,-100:], '.')
plt.xlabel("$\tau$")
plt.ylabel("Calorie Burnage")
plt.show()
```
