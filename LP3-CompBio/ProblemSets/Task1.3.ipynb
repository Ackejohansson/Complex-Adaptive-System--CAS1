```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
alpha = 0.01
R = np.arange(1, 30+.1, .1)
n_gen = 300

eta = np.zeros((len(R), n_gen))
eta_start = 900
eta[:,0] = eta_start
```

```python
def model(eta, R, alpha, n_gen):
    for t in range(n_gen-1):
        eta[:, t+1] = R*eta[:, t]*np.exp(-alpha*eta[:, t])
    return eta
```

```python
eta = model(eta, R, alpha, n_gen)
plt.plot(R, eta[:,-100:])
plt.show()
```

```python
R_four = [10, 20, 30, 40, 50, 60, 70] # TODO: find the correct 4 values


for r in R_four:
    plt.plot(eta[r,:41], label=f'R={R[r]:.1f}')

plt.legend()
plt.title('Task 3b) Four dynamics of population')
plt.xlabel(r'$\tau$')
plt.ylabel(r'$\eta_{\tau}$')
plt.show()
```
